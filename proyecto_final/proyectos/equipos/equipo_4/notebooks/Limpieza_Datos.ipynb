```python
import pandas as pd
import numpy as np
```

```python
# hyperlink to data location
urlData = "/content/healthcare-dataset-stroke-data.csv"

# read in the data to a NumPy array
Data = pd.read_csv(urlData)

# Quitamos valores NA
Data = Data.dropna()

# Impresión para visualización
Data
```

```python
# Limpieza de Datos, volvemos variables que tienen solamente dos nieveles a variables 0-1
Data['gender'] = Data['gender'].replace(['Female','Male', 'Other'],[0,1,0])
Data['ever_married'] = Data['ever_married'].replace(['Yes','No'],[1,0])
Data['Residence_type'] = Data['Residence_type'].replace(['Urban','Rural'],[1,0])
Data["smoking_status"] = Data["smoking_status"].replace(['formerly smoked', 'never smoked', 'smokes', 'Unknown'],[2,0,3,1])

# Cambiamos de posición la variable "stroke", que es la respuesta. 
Data = Data.drop('work_type', 1)
Data = Data.drop('id', 1)
```

```python
cols = Data.columns.tolist()
cols = cols[-1:] + cols[:-1]

Data = Data[cols]
```
