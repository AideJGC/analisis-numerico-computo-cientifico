Empezaremos por hacer una Descomposici√≥n en Valores Singulares, y b√°sada en esta, calcularemos el PCA

```python
dta_clean = pd.read_csv('data/data_clean.csv')

dta_f1 = dta_clean[dta_clean.fecha == 2015]

mtx_f1 = dta_f1.drop(columns = ['fecha', 'estado'])

mtx_clean = dta_clean.drop(columns = ['fecha', 'estado'])
```

```python
U, D, Vh = np.linalg.svd(mtx_f1, full_matrices= False, compute_uv= True)
```

```python
pd.DataFrame(U)
```

El √≠ndice ser√° construido como la primer dimensi√≥n de la matriz

$$U' = UD\Sigma^{-1} $$

Donde $U$ y $D$ se obtienen de la SVD y 

$$\Sigma = \sqrt{eigen(\frac{1}{n-1}X^{T}X)}$$

Donde $X$ es la matriz de datos originales y $n$ es el n√∫mero de renglones de dichos datos

Calculemos, entonces, $\Sigma$:

Primero la matriz de covarianza, es decir, $\frac{1}{n-1}X^{T}X$

```python
X = mtx_f1.to_numpy()

X_t = X.transpose()
```

```python
cov_mtx = X_t@X/31
```

Despu√©s los eigenvalores de dicha matriz

```python
e_vals, e_vects = np.linalg.eig(cov_mtx)
```

Y con esto podemos calcular $\Sigma$ y $\Sigma^{-1}$

```python
sigma = np.sqrt(e_vals)
sigma_inv = 1/sigma
```

As√≠, se tienen todos los elementos para calcular $U'$

```python
U_prime = U@np.diag(D)@np.diag(sigma_inv)
```

```python
pd.DataFrame(U_prime)
```

As√≠, el √≠ndice es la primer columna de $U'$

```python
indice = pd.DataFrame({'estado' : dta_f1['estado'], 'indice' : U_prime[:,0]})

indice.sort_values(['indice'], ascending = False)
```
