## Funciones definidas

```python
def givens_rotation(A):
    (r, c) = np.shape(A)
    Q = np.identity(r)
    R = np.copy(A)
    (rows, cols) = np.tril_indices(r, -1, c)
    for (row, col) in zip(rows, cols):
        if R[row, col] != 0:  # Q = 1, s = 0, r, q sin cambios
            r_ = np.hypot(R[col, col], R[row, col])  # d
            c = R[col, col]/r_
            s = -R[row, col]/r_
            G = np.identity(r)
            G[[col, row], [col, row]] = c
            G[row, col] = s
            G[col, row] = -s
            R = np.dot(G, R)  # R=G(n-1,n)*...*G(2n)*...*G(23,1n)*...*G(12)*A
            Q = np.dot(Q, G.T)  # Q=G(n-1,n).T*...*G(2n).T*...*G(23,1n).T*...*G(12).T
    return (Q, R)
```

```python
#Algoritmo QR para matrices simetricas (version simple)
def algoritmoQR(matrix):
    #T0
    T_k_minus_1 = matrix
    q,r = givens_rotation(matrix)
    m,n = q.shape
    Q = np.eye(m,n)
    T_k = r@q 
    #Q es la matriz que aproximará a los eigenvectores
    Q = Q@q
    n = 100 # Número de iteraciones
    while n > 0:
        T_k_minus_1 = T_k
        q,r = givens_rotation(T_k_minus_1)
        Q = Q@q
        #T_k es la matriz que aproxima a nuestros eigenvalores
        T_k = r@q
        n = n - 1
    
    return np.diag(T_k), Q
```

```python
mat2 = mat.to_numpy()
eigenvalores_u, eigenvectores_u = algoritmoQR(mat2@mat2.T/31) #U
```

Empezaremos por hacer una Descomposición en Valores Singulares, y básada en esta, calcularemos el PCA.

Calculamos $U$

```python
U = pd.DataFrame(eigenvectores_u).iloc[:,0:28]
U
```

El índice será construido como la primer dimensión de la matriz

$$U' = UD\Sigma^{-1} $$

Donde $U$ y $D$ se obtienen de la SVD y 

$$\Sigma = \sqrt{\text{eigen}(\frac{1}{n-1}X^{T}X)}$$

Donde $X$ es la matriz de datos originales y $n$ es el número de renglones de dichos datos

Calculemos, entonces, $\Sigma$ 

```python
sigma = np.sqrt(eigenvalores_u)[:-4]
sigma_inv = 1/sigma
sigma_inv
```

Calculamos $D$

```python
eigenvalores_u, eigenvectores_u = algoritmoQR(mat2@mat2.T) #U
D = np.sqrt(eigenvalores_u)[:-4]
D
```

Verificamos dimensiones

```python
U.shape
```

```python
np.diag(D).shape
```

```python
np.diag(sigma_inv).shape
```

Calculamos $U'$

```python
U_prime = U@np.diag(D)@np.diag(sigma_inv)
```

```python
pd.DataFrame(U_prime)
```

Así, el índice es la primer columna de $U'$

```python
indice = pd.DataFrame({'estado' : df[df.fecha==2015]['estado'], 'indice' : -U_prime.iloc[:,0]})

indice.sort_values(['indice'], ascending = False)
```

## Índice delictivo desde 2015 hasta 2020

```python
ind = pd.DataFrame(index = df[df.fecha==2015]['estado'])
for i in df.fecha.unique():
    print('Seleccionamos año: '+str(i))
    mat = df[df.fecha==i].drop(columns=['fecha','estado'])
    print('Aplicamos función Algortimo QR a datos estandarizados')
    mat2 = mat.to_numpy()
    print('Obtenemos U')
    eigenvalores_u, eigenvectores_u = algoritmoQR(mat2@mat2.T/31) #U
    U = pd.DataFrame(eigenvectores_u).iloc[:,0:28]
    print('Calculamos Sigma inversa')
    sigma = np.sqrt(eigenvalores_u)[:-4]
    sigma_inv = 1/sigma
    print('Calculamos D')
    eigenvalores2, eigenvectores2 = algoritmoQR(mat2@mat2.T) 
    D = np.sqrt(eigenvalores2)[:-4]
    print("Obtenemos U'")
    U_prime = U@np.diag(D)@np.diag(sigma_inv)
    print("Seleccionamos la primera columna de U' como índice")
    print('Almacenamos')
    ind[i] = -U_prime.iloc[:,0].values
    print('***')
```
